```python
import pandas as pd
```

```python
df = pd.read_csv('carprices.csv')
df
```

# **Using Dummy Variables**

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df, dummies], axis = 'columns')
merged
```

```python
final = merged.drop(['Car Model', 'Audi A5'], axis = 'columns')
final
```

```python
X = final.drop(['Sell Price($)'], axis = 'columns').values
X
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```

```python
model.predict([[45000,4,0,1]])
```

```python
model.predict([[86000,7,1,0]])
```

# **Using sklearn OneHotEncoder**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle['Car Model'] = le.fit_transform(dfle['Car Model'])
dfle
```

```python
X = dfle.drop(['Sell Price($)'], axis = 'columns').values
X
```

```python
y = dfle['Sell Price($)'].values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('Sell Price($)', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,45000,4]])
```

```python
model.predict([[1,0,86000,7]])
```
