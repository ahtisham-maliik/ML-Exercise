```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
income = df.drop("year", axis = "columns")
income.head()
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("Per Capita Income (USD)")
plt.scatter(df.year, income, color = "green", marker = "*")
```

```python
year = df.year
year.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["year"]], income)
```

```python
reg.predict([[2020]])
```
